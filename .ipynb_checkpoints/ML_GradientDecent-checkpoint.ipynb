# B√†i t·∫≠p 1: T·ªëi ∆∞u h√≥a h√†m s·ªë 2 bi·∫øn b·∫±ng Gradient Descent

## üéØ M√¥ t·∫£ b√†i to√°n

T·ªëi ∆∞u h√≥a h√†m s·ªë sau b·∫±ng thu·∫≠t to√°n Gradient Descent:

$$
\frac{\partial f}{\partial x} = 2(x - 2), \frac{\partial f}{\partial y} = 2(y + 3)
$$

---

## ‚úèÔ∏è Y√™u c·∫ßu

- T√≠nh to√°n ƒë·∫°o h√†m ri√™ng theo \(x\) v√† \(y\):
  
$$
\frac{\partial f}{\partial x} = 2(x - 2)
$$

$$
\frac{\partial f}{\partial y} = 2(y + 3)
$$

- Kh·ªüi t·∫°o gi√° tr·ªã ban ƒë·∫ßu:
  - \(x = 0\)
  - \(y = 0\)

- √Åp d·ª•ng thu·∫≠t to√°n Gradient Descent:
  - Learning rate = 0.1
  - T·ªëi thi·ªÉu 50 v√≤ng l·∫∑p ho·∫∑c d·ª´ng s·ªõm n·∫øu gradient nh·ªè h∆°n \(1e^{-6}\).

---

## üîç Output y√™u c·∫ßu

- In ra gi√° tr·ªã \(x\), \(y\) v√† \(f(x, y)\) sau m·ªói v√≤ng l·∫∑p.
- G·ª£i √Ω output:

```python
# gradient_descent.py

# Kh·ªüi t·∫°o tham s·ªë
x = 0.0
y = 0.0
learning_rate = 0.1
max_iterations = 50
tolerance = 1e-6

# H√†m t√≠nh gi√° tr·ªã c·ªßa f(x, y)
def f(x, y):
    return (x - 2) ** 2 + (y + 3) ** 2

# H√†m t√≠nh gradient
def compute_gradient(x, y):
    df_dx = 2 * (x - 2)
    df_dy = 2 * (y + 3)
    return df_dx, df_dy

# List l∆∞u k·∫øt qu·∫£
results = []

# V√≤ng l·∫∑p Gradient Descent
for i in range(1, max_iterations + 1):
    grad_x, grad_y = compute_gradient(x, y)
    gradient_norm = (grad_x ** 2 + grad_y ** 2) ** 0.5

    if gradient_norm < tolerance:
        break

    x = x - learning_rate * grad_x
    y = y - learning_rate * grad_y
    value = f(x, y)

    # L∆∞u k·∫øt qu·∫£ m·ªói v√≤ng l·∫∑p v√†o list
    results.append((i, x, y, value))

# In k·∫øt qu·∫£ cu·ªëi c√πng
print(f"\nFinal result: x ‚âà {x:.6f}, y ‚âà {y:.6f}, f(x, y) ‚âà {f(x, y):.6f}")

# Export results ƒë·ªÉ v·∫Ω ·ªü file kh√°c
# (N·∫øu l√† .py ƒë·ªôc l·∫≠p th√¨ c√≥ th·ªÉ d√πng pickle/json, c√≤n trong notebook th√¨ kh√¥ng c·∫ßn)
```

```python
# plot_results.py

import matplotlib.pyplot as plt

# Gi·∫£ s·ª≠ bi·∫øn results ƒë√£ ƒë∆∞·ª£c import ho·∫∑c copy t·ª´ file gradient_descent.py
# results = [(iteration, x, y, value), ...]

# N·∫øu ƒëang l√†m trong c√πng file .ipynb, ch·ªâ c·∫ßn ch·∫°y sau cell 1

# T√°ch d·ªØ li·ªáu
iterations = [row[0] for row in results]
x_values = [row[1] for row in results]
y_values = [row[2] for row in results]
function_values = [row[3] for row in results]

# V·∫Ω bi·ªÉu ƒë·ªì qu√° tr√¨nh h·ªôi t·ª•
plt.figure(figsize=(12, 4))

# Bi·ªÉu ƒë·ªì gi√° tr·ªã f(x, y)
plt.subplot(1, 3, 1)
plt.plot(iterations, function_values, marker='o')
plt.title('Gi√° tr·ªã f(x, y) theo iterations')
plt.xlabel('Iteration')
plt.ylabel('f(x, y)')

# Bi·ªÉu ƒë·ªì gi√° tr·ªã x
plt.subplot(1, 3, 2)
plt.plot(iterations, x_values, marker='x', color='r')
plt.title('Gi√° tr·ªã x theo iterations')
plt.xlabel('Iteration')
plt.ylabel('x')

# Bi·ªÉu ƒë·ªì gi√° tr·ªã y
plt.subplot(1, 3, 3)
plt.plot(iterations, y_values, marker='^', color='g')
plt.title('Gi√° tr·ªã y theo iterations')
plt.xlabel('Iteration')
plt.ylabel('y')

plt.tight_layout()
plt.show()
```

```python
import json
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# ƒê·ªçc d·ªØ li·ªáu
with open('gradient_results.json', 'r') as f_in:
    results = json.load(f_in)

x_values = [row["x"] for row in results]
y_values = [row["y"] for row in results]
z_values = [row["z"] for row in results]

# V·∫Ω 3D
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

ax.plot(x_values, y_values, z_values, marker='o')

ax.set_title("Gradient Descent Trajectory in 3D Space")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x, y)")

plt.show()
```

## üéØ M√¥ t·∫£
